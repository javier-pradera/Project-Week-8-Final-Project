# Modelisation with Machine Learning

In this notebook, we will take the cleaned data as an input, we will preprocess it and train different machine learning algorithms to check which is the best fit.

```python
# I first import the necessary libraries:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# I then import the dataset and create the X and the y:
appointments = pd.read_csv('Data cleaning/Appointments_dataset_clean.csv',index_col=0)
X = appointments.drop('No-show',axis=1)
y = appointments['No-show']
```

```python
X
```

```python
y.value_counts()
```

```python
# I now select the train and test sets:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=123)
```

```python
y_test
```

```python
# I will now preprocess the X for the models:
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## 1. Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV
dec_tree = DecisionTreeClassifier()
dec_tree_search = RandomizedSearchCV(estimator=dec_tree,
                                    param_distributions={'criterion':['gini','entropy'],
                                                        'max_depth':range(2,50)},
                                    scoring=,
                                    n_iter=80)
dec_tree_search.fit(X_train_scaled,y_train)
```

```python
dec_tree_search.best_score_
```

```python
dec_tree_search.best_params_
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
y_pred_tree = dec_tree_search.predict(X_test_scaled)
conf_matr_tree = confusion_matrix(y_true=y_test,y_pred=y_pred_tree)
score_tree = accuracy_score(y_true=y_test, y_pred=y_pred_tree)
print(conf_matr_tree)
print(score_tree)
```
